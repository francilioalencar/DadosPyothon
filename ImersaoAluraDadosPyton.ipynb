```python
import pandas as pd
import plotly.express as px
pd.options.display.float_format= '{:.2f}'.format

```

```python
#Importa os dados para o data frame
df_principal = pd.read_excel('/content/Imersão Python -Tabela-Original.xlsx', sheet_name='Principal')
df_principal
```

```python
df_total_de_acoes = pd.read_excel('/content/Imersão Python -Tabela-Original.xlsx', sheet_name='Total_de_acoes')
df_total_de_acoes
```

```python
df_ticker = pd.read_excel('/content/Imersão Python -Tabela-Original.xlsx', sheet_name='Ticker')
df_ticker
```

```python
df_gpt = pd.read_excel('/content/Imersão Python -Tabela-Original.xlsx', sheet_name='GPT')
df_gpt
```

```python
#Selecionar os dados de trabalho
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()

#Renomeia para retrar caracteres especiais e padroniza com todas as letras em caixa baixa
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Último (R$)': 'ultimo_valor','Var. Dia (%)': 'variacao_dia'})

df_principal
```

```python
#Cria nova coluna no df_principal
df_principal['variacao_percentual'] = df_principal['variacao_dia'] /100

#Cria nova coluna no df_principal
df_principal['valor_inicial'] = df_principal['ultimo_valor'] /(df_principal['variacao_percentual']+1)

df_principal
```

```python
df_principal = df_principal.merge( df_total_de_acoes, left_on='ativo', right_on='Código', how='left' )
df_principal
```

```python
df_principal = df_principal.drop(columns='Código')
df_principal
```

```python
#Renomeia para retrar caracteres especiais e padroniza com todas as letras em caixa baixa
df_principal = df_principal.rename(columns={'Qtde. Teórica':'quantidade_teorica'})


df_principal['quantidade_teorica'] = df_principal['quantidade_teorica'].astype(int)

df_principal
```

```python
#Cria nova coluna no df_principal
df_principal['variacao_em_real'] = (df_principal['ultimo_valor'] - df_principal['valor_inicial']) * df_principal['quantidade_teorica']
df_principal
```

```python

# lambda x: (executa a função linha a linha)
df_principal['resultado'] = df_principal['variacao_em_real'].apply(lambda x: 'Subiu'  if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge( df_gpt, left_on='ativo', right_on='Ativo', how='left' )
df_principal
```

```python
df_principal = df_principal.drop(columns='Ativo')
df_principal
```

```python
# lambda x: (executa a função linha a linha)
df_principal = df_principal.rename(columns={'Idade (anos)': 'idade_em_anos'})
df_principal['categoria_idade'] = df_principal['idade_em_anos'].apply(lambda x: '> 100'  if x > 100 else ('Entre 50 e 100' if x >= 50 else '< 50'))
df_principal
```

```python
maior = df_principal['variacao_em_real'].max()

menor = df_principal['variacao_em_real'].min()

media = df_principal['variacao_em_real'].mean()

media_quem_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_em_real'].mean()

media_quem_desceu = df_principal[df_principal['resultado']== 'Desceu']['variacao_em_real'].mean()

print(f"Maior\tR$ {maior:,.2f}")

print(f"Menor\tR$ {menor:,.2f}")

print(f"Médio\tR$ {media:,.2f}")

print(f"Média de quem subiu\tR$ {media_quem_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_quem_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_em_real'].sum().reset_index()

df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_em_real'].sum().reset_index()

df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_em_real', text='variacao_em_real', title='Variação em reais por resuultado')
fig.show()
```
